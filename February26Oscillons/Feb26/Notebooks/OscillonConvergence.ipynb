## Diagnostic self-convergence: overlap test

Each diagnostic (rho_c, delta_c, C, M, ...) is a scalar time series computed independently at each resolution. No spatial interpolation needed — just compare the time series:
- $|Q_{LR} - Q_{MR}|$ vs $2^p \cdot |Q_{MR} - Q_{HR}|$

If the rescaled curves overlap, the scheme is $p$-th order.

```python
diag_keys = {
    "rho_c":   r"$\rho_c$  (central density)",
    "delta_c": r"$\delta_c$  (density contrast)",
    "C":       r"$\mathcal{C}$  (compactness)",
    "M":       r"$M$  (oscillon mass)",
    "R":       r"$R$  (oscillon radius)",
    "K_avg":   r"$\langle K \rangle$  (volume-averaged K)",
}

fig, axes = plt.subplots(3, 2, figsize=(14, 14))
axes = axes.ravel()

for ax_idx, (key, label) in enumerate(diag_keys.items()):
    Q_LR = osc_LR[key]
    Q_MR = osc_MR[key]
    Q_HR = osc_HR[key]

    diff_LM = np.abs(Q_LR - Q_MR)
    diff_MH = np.abs(Q_MR - Q_HR)

    ax = axes[ax_idx]
    ax.semilogy(t, diff_LM,       'b-',  lw=1.2, label=r'$|Q_{LR} - Q_{MR}|$')
    ax.semilogy(t, 16 * diff_MH,  'r--', lw=1.2, label=r'$16\,|Q_{MR} - Q_{HR}|$  (4th)')
    ax.semilogy(t, 4 * diff_MH,   'g:',  lw=1.0, label=r'$4\,|Q_{MR} - Q_{HR}|$  (2nd)')
    ax.set_xlabel('t')
    ax.set_ylabel('Absolute difference')
    ax.set_title(label)
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)

fig.suptitle('Diagnostic self-convergence  (LR : MR : HR  =  1 : 2 : 4)', fontsize=14)
fig.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

## Measured convergence order vs time

$p(t) = \log_2\!\Big(|Q_{LR} - Q_{MR}|\;/\;|Q_{MR} - Q_{HR}|\Big)$

```python
fig, axes = plt.subplots(3, 2, figsize=(14, 14))
axes = axes.ravel()

for ax_idx, (key, label) in enumerate(diag_keys.items()):
    diff_LM = np.abs(osc_LR[key] - osc_MR[key])
    diff_MH = np.abs(osc_MR[key] - osc_HR[key])

    with np.errstate(divide='ignore', invalid='ignore'):
        order = np.log2(diff_LM / diff_MH)

    valid = (diff_MH > 1e-14) & np.isfinite(order)

    ax = axes[ax_idx]
    ax.plot(t[valid], order[valid], 'k-', lw=1.0)
    ax.axhline(4, color='r', ls='--', lw=0.8, label='4th order')
    ax.axhline(2, color='g', ls='--', lw=0.8, label='2nd order')
    ax.set_xlabel('t')
    ax.set_ylabel('Convergence order p')
    ax.set_title(label)
    ax.set_ylim(-1, 8)
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)

fig.suptitle(r'Measured convergence order  $p = \log_2(|Q_{LR}-Q_{MR}|/|Q_{MR}-Q_{HR}|)$',
             fontsize=13)
fig.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

## Hamiltonian constraint convergence

The constraint should converge to zero at the expected order. We check the L2 norm $\|H\|$ at each resolution.

```python
n_sample = 20
t_sample_idx = np.linspace(0, len(t)-1, n_sample, dtype=int)
t_sample = t[t_sample_idx]

r_min_phys = r_LR[3]
r_max_phys = 0.8 * r_max

# Load one at a time, compute constraint, delete
sol = np.load(os.path.join(DATA_DIR, "solution_LR.npy"))
Ham_LR, _ = get_constraints_diagnostic(sol[t_sample_idx], t_sample,
                                        grid_LR, background_LR, my_matter, ppp)
del sol; gc.collect()

sol = np.load(os.path.join(DATA_DIR, "solution_MR.npy"))
Ham_MR, _ = get_constraints_diagnostic(sol[t_sample_idx], t_sample,
                                        grid_MR, background_MR, my_matter, ppp)
del sol; gc.collect()

sol = np.load(os.path.join(DATA_DIR, "solution_HR.npy"))
Ham_HR, _ = get_constraints_diagnostic(sol[t_sample_idx], t_sample,
                                        grid_HR, background_HR, my_matter, ppp)
del sol; gc.collect()

mask_ham_LR = (r_LR >= r_min_phys) & (r_LR <= r_max_phys)
mask_ham_MR = (r_MR >= r_min_phys) & (r_MR <= r_max_phys)
mask_ham_HR = (r_HR >= r_min_phys) & (r_HR <= r_max_phys)

norm_Ham_LR = np.sqrt(np.mean(Ham_LR[:, mask_ham_LR]**2, axis=1))
norm_Ham_MR = np.sqrt(np.mean(Ham_MR[:, mask_ham_MR]**2, axis=1))
norm_Ham_HR = np.sqrt(np.mean(Ham_HR[:, mask_ham_HR]**2, axis=1))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.semilogy(t_sample, norm_Ham_LR, 'bo-', ms=4, label=f'LR (dr={min_dr_LR})')
ax1.semilogy(t_sample, norm_Ham_MR, 'gs-', ms=4, label=f'MR (dr={min_dr_MR})')
ax1.semilogy(t_sample, norm_Ham_HR, 'r^-', ms=4, label=f'HR (dr={min_dr_HR})')
ax1.set_xlabel('t')
ax1.set_ylabel(r'$\|H\|_{L2}$')
ax1.set_title('Hamiltonian constraint violation')
ax1.legend()
ax1.grid(True, alpha=0.3)

with np.errstate(divide='ignore', invalid='ignore'):
    ratio_LR_MR = norm_Ham_LR / norm_Ham_MR
    ratio_MR_HR = norm_Ham_MR / norm_Ham_HR

ax2.plot(t_sample, ratio_LR_MR, 'bo-', ms=4, label=r'$\|H_{LR}\| / \|H_{MR}\|$')
ax2.plot(t_sample, ratio_MR_HR, 'gs-', ms=4, label=r'$\|H_{MR}\| / \|H_{HR}\|$')
ax2.axhline(16, color='r', ls='--', lw=0.8, label='4th order (16)')
ax2.axhline(4,  color='orange', ls='--', lw=0.8, label='2nd order (4)')
ax2.set_xlabel('t')
ax2.set_ylabel('Ratio')
ax2.set_title('Constraint convergence ratio')
ax2.legend(fontsize=9)
ax2.grid(True, alpha=0.3)

fig.tight_layout()
plt.show()
```

## Relative error of HR solution (Richardson extrapolation)

Using $Q_{\rm exact} \approx Q_{HR} + \frac{Q_{MR} - Q_{HR}}{2^p - 1}$ with $p=4$, we estimate the remaining error in HR.

```python
p_assumed = 4
richardson_factor = 1.0 / (2**p_assumed - 1)  # = 1/15

print(f"Assuming p = {p_assumed},  Richardson correction factor = {richardson_factor:.4f}\n")
print(f"{'Diagnostic':<12s}  {'max |Q_MR-Q_HR|':>16s}  {'max |Q_HR| ':>12s}  {'max rel err HR':>16s}")
print("-" * 65)

for key in ["rho_c", "delta_c", "C", "M", "R", "K_avg"]:
    diff_MH    = np.abs(osc_MR[key] - osc_HR[key])
    abs_HR     = np.abs(osc_HR[key])
    err_HR_est = richardson_factor * diff_MH

    max_diff   = np.max(diff_MH)
    max_abs    = np.max(abs_HR)
    max_rel    = np.max(err_HR_est / np.maximum(abs_HR, 1e-30))

    print(f"{key:<12s}  {max_diff:16.3e}  {max_abs:12.3e}  {max_rel:16.3e}")
```

## Predicted-vs-actual convergence (Ham-style plot)

Plot the difference from HR as a function of $\ln(a)$. If $p$-th order:

$$Q_{LR} - Q_{HR} \approx (Q_{MR} - Q_{HR}) \times \frac{h_{LR}^{\,p} - h_{HR}^{\,p}}{h_{MR}^{\,p} - h_{HR}^{\,p}}$$

With $h_{LR}/h_{HR} = 4$ and $h_{MR}/h_{HR} = 2$, the rescaling factors are $17$ (4th order) and $5$ (2nd order).

```python
ln_a = osc_HR["ln_a"]

# Rescaling factors: (h_LR^p - h_HR^p) / (h_MR^p - h_HR^p)
# h_LR/h_HR = 4, h_MR/h_HR = 2
factor_4th = (4**4 - 1) / (2**4 - 1)   # = 255/15 = 17
factor_2nd = (4**2 - 1) / (2**2 - 1)   # = 15/3  = 5

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ── Central density ──
ax = axes[0]
diff_LR = osc_LR["rho_c"] - osc_HR["rho_c"]
diff_MR = osc_MR["rho_c"] - osc_HR["rho_c"]

ax.plot(ln_a, diff_LR,               '--', color='b',      lw=1.5, label=r'LR $\rho_c$ $-$ HR $\rho_c$')
ax.plot(ln_a, diff_MR,               '-',  color='orange',  lw=1.2, label=r'MR $\rho_c$ $-$ HR $\rho_c$')
ax.plot(ln_a, diff_MR * factor_4th,  ':',  color='green',   lw=1.5, label=rf'(MR$-$HR) $\times$ {factor_4th:.0f}  (4th order)')
ax.plot(ln_a, diff_MR * factor_2nd,  '-.', color='red',     lw=1.0, label=rf'(MR$-$HR) $\times$ {factor_2nd:.0f}  (2nd order)')

ax.set_xlabel(r'$\ln(a)$')
ax.set_ylabel(r'$\Delta\rho_c$')
ax.set_title(r'Central density $\rho_c$')
ax.legend(fontsize=8)
ax.grid(True, alpha=0.3)

# ── Compactness ──
ax = axes[1]
diff_LR = osc_LR["C"] - osc_HR["C"]
diff_MR = osc_MR["C"] - osc_HR["C"]

ax.plot(ln_a, diff_LR,               '--', color='b',      lw=1.5, label=r'LR $\mathcal{C}$ $-$ HR $\mathcal{C}$')
ax.plot(ln_a, diff_MR,               '-',  color='orange',  lw=1.2, label=r'MR $\mathcal{C}$ $-$ HR $\mathcal{C}$')
ax.plot(ln_a, diff_MR * factor_4th,  ':',  color='green',   lw=1.5, label=rf'(MR$-$HR) $\times$ {factor_4th:.0f}  (4th order)')
ax.plot(ln_a, diff_MR * factor_2nd,  '-.', color='red',     lw=1.0, label=rf'(MR$-$HR) $\times$ {factor_2nd:.0f}  (2nd order)')

ax.set_xlabel(r'$\ln(a)$')
ax.set_ylabel(r'$\Delta\mathcal{C}$')
ax.set_title(r'Compactness $\mathcal{C}$')
ax.legend(fontsize=8)
ax.grid(True, alpha=0.3)

fig.suptitle('Convergence: LR difference vs predicted from MR difference', fontsize=13)
fig.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

## Spatial convergence of density profile (paper Fig. C.1)

Pick a time snapshot after the oscillon has formed and compare $\rho(r)$ pointwise across the three resolutions. The "high – medium" error rescaled by $2^p$ (for $p = 2$ and $p = 4$) should bracket the "medium – low" error.

```python
# ── Spatial convergence of rho(r) — reproducing paper Fig. C.1 ──

# Choose a snapshot time (late enough that the oscillon has formed).
# Adjust t_snapshot to a time where rho_c has stabilised.
t_snapshot = 600.0
idx_snap = np.argmin(np.abs(t - t_snapshot))
print(f"Snapshot at t = {t[idx_snap]:.1f}  (index {idx_snap})")

# Extract rho(r) at the snapshot for each resolution
rho_snap_LR = osc_LR["rho"][idx_snap]
rho_snap_MR = osc_MR["rho"][idx_snap]
rho_snap_HR = osc_HR["rho"][idx_snap]

# Interpolate LR and MR profiles onto the HR grid for pointwise comparison
from scipy.interpolate import interp1d

rho_LR_on_HR = interp1d(r_LR, rho_snap_LR, kind='cubic',
                         fill_value='extrapolate')(r_HR)
rho_MR_on_HR = interp1d(r_MR, rho_snap_MR, kind='cubic',
                         fill_value='extrapolate')(r_HR)
rho_HR_on_HR = rho_snap_HR

# Differences
diff_ML = np.abs(rho_MR_on_HR - rho_LR_on_HR)   # "medium – low"
diff_HM = np.abs(rho_HR_on_HR - rho_MR_on_HR)    # "high – medium"

# Restrict to the inner region where the oscillon lives
r_plot_max = 30.0
pmask = (r_HR > 0.5) & (r_HR < r_plot_max)

# ── Two-panel figure matching Fig. C.1 ──
fig, (ax_top, ax_bot) = plt.subplots(2, 1, figsize=(8, 7),
                                      sharex=True,
                                      gridspec_kw={'height_ratios': [1, 1.3]})

# Top panel: density profile at HR
ax_top.plot(r_HR[pmask], rho_HR_on_HR[pmask], 'k-', lw=1.5)
ax_top.set_ylabel(r'$\rho \;\;[m^2 M_{\rm Pl}^2]$', fontsize=12)
ax_top.set_title(rf'Oscillon density profile at $t = {t[idx_snap]:.0f}$',
                 fontsize=13)
ax_top.ticklabel_format(axis='y', style='sci', scilimits=(-4,-4))
ax_top.grid(True, alpha=0.3)

# Bottom panel: log(error) with 2nd-to-4th order band
floor = 1e-30
log_diff_ML = np.log10(np.maximum(diff_ML[pmask], floor))
log_diff_HM = np.log10(np.maximum(diff_HM[pmask], floor))

# Rescale high–medium by 2^p for p = 2 (factor 4) and p = 4 (factor 16)
log_HM_2nd = np.log10(np.maximum(4  * diff_HM[pmask], floor))
log_HM_4th = np.log10(np.maximum(16 * diff_HM[pmask], floor))

ax_bot.fill_between(r_HR[pmask], log_HM_2nd, log_HM_4th,
                    color='limegreen', alpha=0.4,
                    label=r'$2^{\rm nd}$ to $4^{\rm th}$ order')
ax_bot.plot(r_HR[pmask], log_diff_ML, 'r-',  lw=1.5,
            label='medium $-$ low')
ax_bot.plot(r_HR[pmask], log_diff_HM, 'b--', lw=1.5,
            label='high $-$ medium')

ax_bot.set_xlabel(r'$r \;\;[1/m]$', fontsize=12)
ax_bot.set_ylabel(r'$\log_{10}({\rm error}[\rho])$', fontsize=12)
ax_bot.legend(fontsize=10, loc='lower right')
ax_bot.grid(True, alpha=0.3)

fig.tight_layout()
plt.show()

print(f"\nResolution:  LR dr={min_dr_LR},  MR dr={min_dr_MR},  HR dr={min_dr_HR}")
print(f"Refinement ratio  h_LR/h_MR = {min_dr_LR/min_dr_MR:.0f},  "
      f"h_MR/h_HR = {min_dr_MR/min_dr_HR:.0f}")
```
